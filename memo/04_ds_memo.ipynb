# Stakeholder

This analysis is intended for **state legislators (House and Senate) and Governors** responsible for designing, approving, and funding correctional policies.

These stakeholders determine:
- The structure and funding of prison education programs  
- The adoption of policies such as sentence reduction incentives, automatic enrollment, and centralized administration  
- The broader strategy for reducing recidivism and improving post-release outcomes  

The goal of this memo is to summarize the updated evidence on education policies and recidivism, with clear guidance on what is supported, what is uncertain, and what should be evaluated next.

________

# Executive Summary

The United States incarcerates approximately 1.9 million individuals, with over 1.25 million in state and federal prisons, the highest rate globally. High recidivism sustains economic and social strain on communities, families, and correctional systems. Education programs are among the most studied tools for reducing reoffending, but the evidence depends on both access and policy design.

This analysis uses 11.5 million individual prison releases across 34 states from 2005 to 2018, drawn from the National Corrections Reporting Program (NCRP). It examines two pathways: expanded access to college education through the 2016 Second Chance Pell initiative, and cross-state variation in the quality of prison education policy design.

**Pell access.** Pell states experienced a raw 12.2 percentage point decline in two-year recidivism after 2016, compared to a 6.2 point decline in non-Pell states , a raw difference of roughly 6 percentage points. After adding state fixed effects, year effects, and individual controls with state-clustered standard errors, the preferred estimate is a **2.4 percentage point reduction**, which is **not statistically significant** (p = 0.25). The signal is directionally favorable but not conclusive.

**Policy design.** Cross-sectional models relating a composite state policy strength score, and its seven components , to recidivism find **no statistically significant associations** across any specification. With only 33 states providing policy data, the analysis is underpowered for this type of cross-sectional identification.

The practical implication: expand and sustain education access, but pair it with rigorous evaluation. Policy design effects are not yet clearly detectable in current data.

# Policy Recommendations

### 1. Maintain and Evaluate Pell Access
- Continue supporting access to postsecondary education through Pell-funded programs  
- The updated estimates point to a **directionally lower** recidivism rate, but the preferred model is **not statistically significant**, so continued evaluation is essential  
- Pair program expansion with outcome tracking to identify where and for whom benefits are strongest  

### 2. Invest in Implementation Quality and Participation
- Access alone is unlikely to be sufficient without consistent delivery and participation  
- Strengthen program availability across facilities, reduce waitlists, and monitor completion rates  
- Track enrollment, attendance, and credential outcomes alongside recidivism  

### 3. Pilot Incentive-Based Policies with Clear Evaluation
- Incentive structures are plausible but the current data do **not** show a robust effect  
- Use small-scale pilots with predefined outcomes rather than broad rollout  
- Prioritize designs that can be evaluated using pre-registered metrics and comparison groups  

### 4. Improve Data for Policy Design Decisions
- Current policy indicators are coarse and may not capture implementation intensity  
- Collect program-level measures such as instructional hours, credential completion, and staffing ratios  
- Better data will allow future analyses to distinguish which policy designs actually improve outcomes

# Context and Background

Prison education programs have evolved from 18th-century religious instruction and basic literacy into a broad range of vocational training, secondary education, and postsecondary coursework. A significant shift occurred in 1994, when Pell Grant eligibility for incarcerated individuals was eliminated, causing a sharp decline in college programs nationwide. In 2015–2016, the Second Chance Pell initiative restored access on an experimental basis in participating states, representing the most substantial policy change in prison education in two decades.

The broader evidence base largely supports education as a tool for rehabilitation. Studies consistently find that participants in prison education programs are less likely to return to prison, though separating program effects from selection - participants may be more motivated to begin with - remains a methodological challenge. This analysis uses the policy change in 2016 as a natural experiment to partially address selection concerns.

Beyond access, how states structure prison education varies considerably: some require automatic enrollment, some attach sentence reductions to program completion, and some operate programs through centralized school districts. Whether these structural differences predict recidivism outcomes is the second question this analysis addresses.

# Data and Approach

This analysis uses individual-level data from the National Corrections Reporting Program (NCRP), covering 11.5 million prison releases across 34 states. The analysis window is 2005–2018. Recidivism is defined as a return to prison within two years of release, measured using each individual's subsequent admission record. The overall two-year recidivism rate in the analysis sample is 26.3%.

Two complementary analyses are used:

**Pell difference-in-differences.** We compare recidivism trends in 19 Pell-participating states before and after 2016 against trends in 15 non-participating states. State fixed effects absorb persistent differences between states; year fixed effects absorb national trends. Individual controls include offense type, sex, age at admission, and sentence length. Standard errors are clustered by state, since the treatment varies at the state level.

**Policy design cross-section.** We relate a composite state policy strength score (from a 2024 study of state prison education laws) and its seven components to recidivism. These scores do not vary over time, so all identifying variation is across states. With 33 states providing policy data, this is a cross-sectional comparison of roughly 33 units, regardless of individual-level sample size. State fixed effects cannot be included without absorbing all policy variation.

## Methodology 1: Pell Second Chance Program

To evaluate the impact of expanded access to education, this analysis examines the Second Chance Pell initiative, which restored access to college-level education for incarcerated individuals beginning in 2016.

The approach compares recidivism changes in Pell states relative to non-Pell states, controlling for state and year differences and individual characteristics.

This provides a structured test of whether Pell access is associated with lower recidivism, while acknowledging that other policy changes may also influence trends.

![Pell Recidivism Plot](../plots/pell_pre_post_bars.png)

The raw comparison is striking: Pell states saw recidivism fall from 28.4% to 16.2% after 2016, while non-Pell states fell from 25.3% to 19.1% - a raw difference-in-differences of about 6 percentage points. However, some of this decline reflects national trends (visible in both groups) and composition differences between states.

The preferred model , with state fixed effects, year fixed effects, individual controls, and state-clustered standard errors , estimates a **2.4 percentage point reduction** in recidivism for Pell states, and this estimate is **not statistically significant** (p = 0.25, 95% CI: −6.5 to +1.7 pp).

This means the Pell signal is **directionally favorable but not conclusive**. The estimate shrinks when national trends and persistent state differences are controlled for. Additionally, the post-2016 period overlaps with other criminal justice reforms, so the estimate cannot be attributed solely to Pell expansion.

## Methodology 2: State Policy Design

This section asks whether states with stronger prison education policies have lower recidivism. A composite score from a 2024 study rates each state across seven dimensions: adult basic education, secondary education, vocational programs, college programs, automatic enrollment, school district structure, and sentence reduction incentives.

**The results show no statistically significant association** between policy strength and recidivism , neither the composite score nor any individual component reaches conventional significance across individual-level or state-year weighted models.

Three interpretations are consistent with this null result:
1. **Underpowered.** Cross-sectional identification across 33 states cannot reliably detect a modest effect, even with millions of individual observations. The effective sample for inference is the number of state clusters.
2. **Measurement error.** The 2024 policy scores may not accurately represent each state's policies during 2005–2018. Mismeasurement of this kind attenuates estimates toward zero.
3. **No detectable effect.** State-level policy quality, as scored here, may not independently predict recidivism after accounting for national year trends.

These three explanations cannot be distinguished with the current data. The null should be interpreted as inconclusive, not as evidence that policy design is irrelevant.

# Limitations

**Right-censoring.** Individuals released in 2017–2018 had limited time to return to prison before the data end. The preferred models include year fixed effects that absorb most of this, but post-2016 recidivism estimates should be interpreted with some caution.

**Policy score timing.** The 2024 policy scores represent states' current education law quality, not their policies during 2005–2018. States that reformed their programs over this period are misrepresented in the cross-sectional analysis, biasing estimates toward zero.

**Cross-sectional identification.** The policy design analysis cannot rule out omitted state-level confounders , states with stronger education policies may differ on incarceration philosophy, demographics, or funding in ways that also affect recidivism.

**Confounded post-2016 period.** The Pell expansion coincided with broader criminal justice reforms. Changes in recidivism after 2016 cannot be attributed solely to education access.

**Program quality vs. policy presence.** A state can adopt a policy without implementing it well. The current data measures policy presence, not program intensity, which weakens the link between policy scores and outcomes.

# Conclusion

Across 11.5 million individual releases and 34 states, this analysis finds a directionally favorable but statistically inconclusive association between the 2016 Pell expansion and reduced two-year recidivism. The preferred estimate, 2.4 percentage points, p = 0.25 , shrinks substantially from the raw 6 percentage point gap once national trends and state differences are controlled for. The policy design analysis, limited to 33-state cross-sectional variation, finds no significant associations for any policy component.

The honest summary is that this data does not produce a definitive verdict. The Pell signal is too uncertain to cite as proof of effect, but it is consistent enough to justify sustained investment. The policy design null is too underpowered to rule out real effects.

**The practical path forward is clear regardless of these uncertainties:** maintain and expand education access, build in outcome tracking, and invest in the data infrastructure needed to identify what works and for whom. The question is not whether to fund education in prisons , the directional evidence, combined with the broader literature, supports it. The question is how to design programs and measure them well enough to improve over time.
